# 2. Ventas por categoría

```python
# Normalizar nombres de columnas
tienda.columns = tienda.columns.str.strip().str.lower()
tienda2.columns = tienda2.columns.str.strip().str.lower()
tienda3.columns = tienda3.columns.str.strip().str.lower()
tienda4.columns = tienda4.columns.str.strip().str.lower()

# Contar cantidad de ventas por categoría
categorias_t1 = tienda["categoría del producto"].value_counts()
categorias_t2 = tienda2["categoría del producto"].value_counts()
categorias_t3 = tienda3["categoría del producto"].value_counts()
categorias_t4 = tienda4["categoría del producto"].value_counts()

# Mostrar resultados
print("Tienda 1:\n", categorias_t1, "\n")
print("Tienda 2:\n", categorias_t2, "\n")
print("Tienda 3:\n", categorias_t3, "\n")
print("Tienda 4:\n", categorias_t4, "\n")

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Contar categorías por tienda
cat_t1 = tienda["categoría del producto"].value_counts()
cat_t2 = tienda2["categoría del producto"].value_counts()
cat_t3 = tienda3["categoría del producto"].value_counts()
cat_t4 = tienda4["categoría del producto"].value_counts()

# Unir todas las series en un DataFrame
df_categorias = pd.DataFrame({
    "Tienda 1": cat_t1,
    "Tienda 2": cat_t2,
    "Tienda 3": cat_t3,
    "Tienda 4": cat_t4
}).fillna(0)  # Llenar con 0 las categorías que no existen en alguna tienda

# Gráfico de barras apiladas
df_categorias.plot(kind="bar", stacked=True, figsize=(12,6))
plt.title("Distribución de Categorías por Tienda")
plt.ylabel("Cantidad de Ventas")
plt.xlabel("Categoría")
plt.xticks(rotation=45)
plt.legend(title="Tienda")
plt.tight_layout()
plt.show()

```

# 3. Calificación promedio de la tienda


```python
# Normalizar nombres de columnas
tienda.columns = tienda.columns.str.strip().str.lower()
tienda2.columns = tienda2.columns.str.strip().str.lower()
tienda3.columns = tienda3.columns.str.strip().str.lower()
tienda4.columns = tienda4.columns.str.strip().str.lower()

# Calificación promedio por tienda
calif_tienda1 = tienda["calificación"].mean()
calif_tienda2 = tienda2["calificación"].mean()
calif_tienda3 = tienda3["calificación"].mean()
calif_tienda4 = tienda4["calificación"].mean()

# Resumen en DataFrame
import pandas as pd

resumen_calif = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Calificación Promedio": [calif_tienda1, calif_tienda2, calif_tienda3, calif_tienda4]
})

print(resumen_calif)
```

```python
calificaciones = [calif_tienda1, calif_tienda2, calif_tienda3, calif_tienda4]

plt.figure(figsize=(8,5))
plt.plot(tiendas, calificaciones, marker='o', linestyle='-', color='green')
plt.title("Calificación Promedio por Tienda")
plt.ylabel("Calificación Promedio")
plt.xlabel("Tienda")
plt.ylim(0,5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Normalizar nombres de columnas
tienda.columns = tienda.columns.str.strip().str.lower()
tienda2.columns = tienda2.columns.str.strip().str.lower()
tienda3.columns = tienda3.columns.str.strip().str.lower()
tienda4.columns = tienda4.columns.str.strip().str.lower()

# Función para productos más y menos vendidos
def productos_extremos(df, tienda_nombre):
    ventas_por_producto = df["producto"].value_counts()
    mas_vendido = ventas_por_producto.head(5)
    menos_vendido = ventas_por_producto.tail(5)

    print(f"--- {tienda_nombre} ---")
    print("Productos más vendidos:\n", mas_vendido, "\n")
    print("Productos menos vendidos:\n", menos_vendido, "\n")

# Aplicar a cada tienda
productos_extremos(tienda, "Tienda 1")
productos_extremos(tienda2, "Tienda 2")
productos_extremos(tienda3, "Tienda 3")
productos_extremos(tienda4, "Tienda 4")

```

Tras evaluar todos los indicadores:

La Tienda 1 lidera en ingresos y productos más vendidos, mostrando un fuerte desempeño comercial.

Aunque su costo de envío es mayor, la alta demanda y volumen de ventas lo compensan.

Las categorías más populares se encuentran en esta tienda, asegurando un flujo constante de ventas.

Recomendación: Se sugiere que el Sr. Juan enfoque sus ventas en la Tienda 1, ya que combina alto volumen de ingresos, productos populares y una buena calificación de clientes. Las fortalezas de esta tienda superan sus debilidades, posicionándola como la opción más rentable y estratégica.
